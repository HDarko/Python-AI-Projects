```python
#Sliding Puzzle State Generator


#During a search a great number of 
#board configurations may be made so it is in best interest to make these configurations
#as simple as possible thus they will be strings
#Every 4 numbers represent a single row with numbers above 9 being represented with hexadecimals
# blank space or tile is represented with '_'
#To create a random board state that is solvable we first build a initial completed board
#then make a number of random moves so that we know that the board is solvable
#Also doing this allows me to reuse the code for the testing agent
import random

# This function takes in the size of the board (11 or 15), the number of
# random moves to be made on the initial board state and the number of 
# board states to be generated then outputs them to a file called
#size + 'final' + number of rand moves + .txt
def BoardStateGen(size,ranMoves,states):
    output=""
    if (size==11):
        state="123456789ab_"
        #output= open("11puzzles.txt","w")
        output= open("11final" +str(ranMoves)+".txt","w")
    if(size==15):
        state="123456789ab_"
        output= open("11puzzles.txt","w")
    #print( radMoves(state,ranMoves))
    for i in range(0,states):
        output.write(radMoves(state,ranMoves)+"\n")
    output.close()
        
    

#Function makes number of randMoves as specified on the input board state then returns the result string.
def radMoves(state,randMoves):
    newstate=list(state)
    #add old memory check
    siz=len(newstate)-1
    lastmove=None
    for i in range(randMoves):
        blank=newstate.index("_")
        movelist=pMoves(state,blank,siz)
        if(lastmove==None):
            lastmove=newstate.index("_")
            move=random.choice(movelist)
        else:
            if(lastmove in movelist):
                movelist.remove(lastmove)
            move=random.choice(movelist)
            lastmove=newstate.index("_")
        newstate[blank],newstate[move]=newstate[move],newstate[blank]
    return "".join(newstate)
      
        

 
    #Function that takes the input board state, 
    #the location of blank as an postion index in string and size of the puzzle
    # then returns a list of posible indexs or tile postions the blank can move to.
def pMoves(state,blank,size):
    moves=[]
    #print(state)
    #print(blank)
    if(size==11):
        #if blank can move right then append that position
        #maybe change to even number with division
        if(((blank+1)!=4) and ((blank+1)!=8) and ((blank+1)!=12)):
            #moves.append((blank,state[blank+1]))
            moves.append(blank+1)
        #if the blank can move left get the move
        if(((blank-1)!=(-1)) and ((blank-1)!=3)and((blank-1)!=7)):
            #moves.append((blank,state[blank-1])) 
            moves.append(blank-1)
        #if the blank can move up then get the move
        if((blank+1)>=4):
            #moves.append((blank,state[blank-4]))
            moves.append(blank-4)
        #if the blank can move down then getfor that move
        if((blank+1)<=8):
        
           moves.append(blank+4)
        return moves
    if(size==15):
        #if blank can move right then get that move
        if(((blank+1)!=4) and ((blank+1)!=8) and ((blank+1)!=12) and ((blank+1)!=16)):

            moves.append(blank+1)
        #if the blank can move left get  that move
        if(((blank-1)!=-1) and ((blank-1)!=3)and((blank-1)!=7) and((blank-1)!=11)):
            moves.append(blank-1)
        #if the blank can move up then get that move
        if((blank+1)>=4):
            moves.append(blank-4)
        #if the blank can move down then get that move
        if((blank+1)<=12):
            moves.append(blank+4)
        return moves
               
  
#
#Sample of puzzle board state

#1234
#5678
#9abc
#def_

#First get 3000 states for the 45 random moves puzzles
#To use for training.
#Then create a secondary file for testing with a given number of states
#with a chosen number of random moves

#max number of moves fo 11 puzzle is 45, for 15 is 80
#Code to generate training files for NN
#BoardStateGen(11,42,3000)
#Cod eot generate any file you wish
BoardStateGen(11,20,5)
#BoardStateGen(15,80,3000)
```
